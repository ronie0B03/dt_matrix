# 0. Import Libraries

```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
```

# 1. Read the the dataset

```python
DATA_CSV_FILE = pd.read_csv('sample_data_first-year.csv')
#DATA_CSV_FILE
```

```python
#DATA_CSV_FILE.isnull().sum()
```

# 2. Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10,10)})
sns.lineplot(data=DATA_CSV_FILE, x="PROGRAM", y="TEST")
sns.scatterplot(data=DATA_CSV_FILE, x="PROGRAM", y="TEST")
plt.show()
```

# 3. Preparing Data for training and testing

```python
X = pd.DataFrame(np.c_[DATA_CSV_FILE['TEST'],DATA_CSV_FILE['IT102'], DATA_CSV_FILE['PE10'], DATA_CSV_FILE['NSTP10'],DATA_CSV_FILE['STS101']], columns=['TEST','IT102','PE10','NSTP10', 'STS101'])
y = DATA_CSV_FILE['PROGRAM']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X
```

```python
clf = DecisionTreeClassifier()
```

# 4. Fit the model

```python
clf.fit(X_train, y_train)
```

# 5. Accuracy of the model

```python
print(clf.score(X_test, y_test))
```

# 3. Predict sample input

```python
example_input = [[90,1,1.5,1.25,2],[80,1,1,1,1]]
predict = clf.predict(example_input)
print(predict)
```
